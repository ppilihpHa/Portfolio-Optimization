## Ziel der Arbeit

=> Klassische Portfolio Optimierung ist mathematisch elegegant, aber kann das Ganze auch so in der Praxis umgesetzt werden? Diese Limitationen sollen untersucht werden.

**Konkret:**

=> Portfolio-Optimierung als sensitivitätsgetriebenes Optimierungsproblem unter unsicheren Inputs -> Probleme einorden (am besten quantifizieren) und behandeln

## Forschungsfrage

**RQ**: Wie sensitiv sind klassische Ansätze der Portfolio Optimisierung gegenüber mit der Praxis einhergehenden Problemen?

*Um das Ganze zu beantworten möchte ich folgene Unterfragen beantworten:*

**rq1**: Wie robust sind die Ansätze gegenüber Schätzfehlern?

**rq2**: Welche Unterschiede zwischen In-Sample und Out-of-Sample Ergebnissen gibt es und wie entstehen sie?

**rq3**: Welchen Einfluss hat die Kovarianzmatrix und die Schätzung der Kovarianzmatrix auf die Verteilung der Portfolio-Gewichte?

**rq4**: Wie sensitiv reagiert die Gewichtung des Porfolios auf Änderungen in den Inputs? (nstocks, Ausprägung t, Kapitalgröße)

**rq5**: Welche Ansätze verbessern praktische Nutzbarkeit hinsichtlich Stabilität?

****

*Wie möchte ich diese Fragen beantworten?*

|rq|Experiment|Ziel|
|---|---|---|
|*rq1*|Ergebnis-Abweichungen von marginalen Input-Änderungen messen & Schätzungen gegen tatsächliche Zukunftswerte evaluieren |messen wie akkurat und sensitiv die Schätzungen sind|
|*rq2*|Theorie: Schätzqualität von Sample Schätzern sowie Vergleich von i-S vs. o-o-S Strategien| Probleme von in-Sample quantifizieren, Verbesserungen aber auch Probleme von o-o-S quantifizieren|
|*rq3*|Ergebnisse in Kovarianzschätzern vergleichen sowie Ergebnisse mit verschiedenen Kovarianzen vergleichen| Sensitivität gegenüber Kovarianz quantifizieren|

|*rq5*|Implikationen aus den Ergebnissen ableiten und umsetzten|Stabilere Portfolio Bildung |

## Hauptexperiment

#### Grundaufbau

**Zeitlicher Rahmen**: $T = 20$ Jahre

**Insgesamtes Aktien-Universum**: $N = 5800$ Aktien

**Portfolio Aktien-Universum**: $n_1 = 10, n_2 = 20, n_3 = 30$ 

**Vorgehen**: lookback Rabalancing

**Lookback Fenster**: $2$ Jahre

**Haltezeit**: $1$ Monat, $21$ Handelstage

**Metriken zur Bestimmung des Portfolio-Universums**: $\frac{r - r_{ex}}{risk}$ -> Sharpe (TODO/OPEN, vllt noch andere?)

**Investment Strategien**: $2$ Ziele mit $2$ Methoden $+$ $1$ Baseline 

**Strategien**: $MinVar$ (in_sample und out_of_sample), $MaxSharpe$ (in_sample und out_of_sample), $Naive$ (Baseline)

**Realistische Umgebung**: Transaktionskosten (Broker Fee + Slippage) $+$ constraints (long only, Vollinvestition)

**Auswertungs-Metriken**: $mean$, $std$, $cumulative wealth$, $drawdown$, $turnover$´

**Auswertungs-Bereiche**: Robustheit/Stabilität, Sensitivität, Schätzfehler

**Evaluierung**: Vergleich von Auswertungsmetriken für Grundergebniss, dann Variationsanalyse um Ursprung zu untersuchen -> Wenn Ursprung messbar, Lösungsansätze untersuchen

#### Ablauf

*Generell sollen verschiedene klassische ``Strategien`` und ``Einflussfaktoren`` analysiert werden, indem sie in einer Simulation variiert werden und so die Unteschiede in den Ergebnissen untersucht werden können. Eine Simulation umfasst einen Investmentzeitraum von ``20 Jahren``, worin eine Strategie mit bestimmten zusätzlichen Faktoren verfolgt wird. Jeden Monat werden basierend auf den ``täglichen Daten`` der vergangenen ``2 Jahre`` bestimmt ``viele Aktien`` anhand einer ``Metrik`` ausgewählt, dann wird anhand eines ``Schätzansatzes`` und einer ``Strategie`` ein Portfolio konstruiert und gehalten. Die Ergebnisse werden gespeichert und anschließend untersucht.*

Die markierten Themen sollen angepasst werden, damit die Auswirkungen untersucht werden können.
