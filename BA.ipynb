## Ziel der Arbeit

=> Klassische Portfolio Optimierung ist mathematisch elegegant, aber kann das Ganze auch so in der Praxis umgesetzt werden? Diese Limitationen sollen untersucht werden.

**Konkret:**

=> Portfolio-Optimierung als sensitivitätsgetriebenes Optimierungsproblem unter unsicheren Inputs -> Problem der Schätzunsicherheit einorden (am besten quantifizieren) und behandeln

## Forschungsfrage

**RQ**: Wie werden klassische Ansätze der Portfolio Optimierung von den mit der Praxis einhergehenden Problemen bezüglich der Schätzung verwendeter Daten beeinflusst?

*Um das Ganze zu beantworten möchte ich folgene Unterfragen beantworten:*

**rq1**: Wie unterscheiden sich In-Sample- und Out-of-Sample-Ergebnisse hinsichtlich Performance und Stabilität?

**rq2**: Wie sensitiv reagiert die Portfolio-Gewichtung auf die Schätzunsicherheit der Kovarianzmatrix?

**rq3**: Wie beeinflussen Eigenschaften der Inputs die Optmierung?

**rq4**: Welche Verfahren verbessern die Robustheit der resultierenden Portfolio-Gewichte? 

****

*Wie möchte ich diese Fragen beantworten?*

|rq|Experiment|Ziel|
|---|---|---|
|*rq1*|Schätzungen gegen tatsächliche Werte evaluieren|messen wie akkurat die Schätzungen sind und wie sie das Verhalten verändern|
|*rq2*|Veränderung durch marginale Anpassung der Kovarianzmatrix messen|Sensitivität auf Kovarianz-Schätzunsicherheit quantifizieren|
|*rq3*|Vergleich von Ergebnissen nach Parameter(Portfoliogröße, Fenstergröße)-Variation|Effekt von realistischen Parametern auf Schätzunsicherheit|
|*rq4*|Analyse der gewonnen Ergebnisse|Stabilere Portfolio Bildung|

<br>

****

*Dabei möchte ich folgendes untersuchen:*

|Hypothese|Erwartung|
|:---|:---|
|H1|mathematisch optimierte Strategien erziehlen bessere Performance aber sind instabiler|
|H2|Shrinkage reduziert Instabilität|
|H3|Hohe Instabilität (hoher Turnover) korreliert mit schlechterer Out-of-Sample-Performance|
|H4|Größere Portfolios sind stabiler|
|H5|Feingranularere Inputs erzeugen mehr Noise|


## Hauptexperiment

#### Grundaufbau

**Zeitlicher Rahmen**: $T = 20$ Jahre

**Insgesamtes Aktien-Universum**: $N = 5800$ Aktien

**Portfolio Aktien-Universum**: $n_1 = 10, n_2 = 20, n_3 = 30$ 

**Vorgehen**: lookback Rebalancing

**Lookback Fenster**: $2$ Jahre

**Haltezeit**: $1$ Monat, $21$ Handelstage

**Metriken zur Bestimmung des Portfolio-Universums**: $\frac{r - r_{ex}}{risk}$ -> Sharpe (TODO/OPEN, vllt noch andere?)

**Investment Strategien**: $2$ Ziele mit $2$ Methoden $+$ $1$ Baseline 

**Strategien**: $MinVar$ (Sample und Shrinkage), $MaxSharpe$ (Sample und Shrinkage), $Naive$ (Sample, Baseline)

**Realistische Umgebung**: Transaktionskosten (Broker Fee + Slippage) $+$ constraints (long only, Vollinvestition)

**Auswertungs-Metriken**: $mean$, $std$, $cumulative wealth$, $drawdown$, $turnover$´

**Auswertungs-Bereiche**: Robustheit, Stabilität, Sensitivität, Schätzfehler

**Evaluierung**: Vergleich von Auswertungsmetriken für Grundergebniss, dann Variationsanalyse um Ursprung zu untersuchen -> Wenn Ursprung messbar, Lösungsansätze untersuchen

#### Ablauf

*Generell sollen verschiedene klassische ``Strategien`` und ``Einflussfaktoren`` analysiert werden, indem sie in einer Simulation variiert werden und so die Unteschiede in den Ergebnissen untersucht werden können. Eine Simulation umfasst einen Investmentzeitraum von ``20 Jahren``, worin eine Strategie mit bestimmten zusätzlichen Faktoren verfolgt wird. Jeden Monat werden basierend auf den ``täglichen Daten`` der vergangenen ``2 Jahre`` bestimmt ``viele Aktien`` anhand einer definierten Metrik ausgewählt, dann wird anhand eines ``Schätzansatzes`` und einer ``Strategie`` ein Portfolio konstruiert und gehalten. Die Ergebnisse werden gespeichert und anschließend untersucht.*

Die markierten Themen sollen angepasst werden, damit die Auswirkungen untersucht werden können.
