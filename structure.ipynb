# Introduction to Portfolio Optimization - handling practical issues

### Important Concepts from the book

- Linear and Log **returns**
    
    - linear return: $r^{lin}_t = \dfrac{p_t - p_{t-1}}{p_{t-1}}$

    - log return: $r^{log}_t = \log p_t - \log p_{t-1}$

    - since returns are rather small the book assumes: $ \mu^{lin}_t \approx \mu^{log}_t $ and therfore $ \Sigma^{lin}_t \approx \Sigma^{log}_t $ 

- **mean vector** and **covariance matrix**

    -  <font color='green'> mean vector </font> $\color{green}{\mu_t}:$

        - $ \mu^{log}_t = \mathrm{E}[r^{log}_t | \mathcal{F}_{t-1} ] $
        - which is basically the expected return of all assets given all information (forecast) up to $t-1$
    
    - <font color='green'> covarianve matrix </font> $\color{green}{\Sigma_t}:$

        - $ \Sigma_t = [ (r^{log}_t - \mu^{log}_t) (r^{log}_t - \mu^{log}_t) | \mathcal{F}_{t-1} ] $
        - which is basically the expected variance of a assent on the diagonal an the expected covariance on the off-diagonal, given all information (forecast) up to $t-1$

- **Portfolio Return** and **Net Asset Value (NAV)**

    - $w_t$ <font color='red'> portfolio vector, e.q. units of assetes held </font>

        - if units of assets stay constant, then $w_t$ changes with price changes $p_t$

    - $NAV = w^{cap}_t + c^{cap}_t$ <font color='red'> current portfolio value + current cash held  </font>

    - $R^{portfolio}_t = \dfrac{NAV_t - NAV_t-1}{NAV_t-1}$ <font color='red'> currrent return of the portfolio </font>

        - $= w^T_{t-1} * r_t$  <font color='red'> prtfolio vector times individual returns </font>

- first- an second-order **moments** of Potfolios 

    - **Expected portfolio return** $ = w^T_{t-1} \mu $

    - **Portfolio Variance** $ = w^T_{t_1} \Sigma w_{t-1} $ 

    - <font color='red'>Recap:</font> $w_t$ or the weights represent the portfolio (e.g. units held)

- **Cumulative Return**

    - normalized $NAV$ 

- **Transaction Cost**

    - The cost associated with a finacial trade
    
    - *commission fee*: payment to the broker <font color='red'>(dependent on country, broker, etc.)</font>

    - *slippage*: difference between the expected price and the actually executed price <font color='red'>(not necessarily negative)</font>

    - $ (w^{cap}_t \to w^{cap,reb}_t) $ <font color='red'>transaction cost when rebalancing portfolio</font>

    - Therfore the return of the rebalanced portfolio: $ R^{portf}_{t-1} = (w^{reb}_{t-1})^T r_t - (w^{cap}_t \to w^{cap,reb}_t)$

- **Rebalancing**

    - <font color='red'> Trade-off: </font> *frequent rebalancing* to match the currently desired portfolio *vs.* *transaction cost* of rebalancing

    - <font color='red'> Adaptive rebalancing: </font> only rebalance if the *distance* is below some threshold ($\to$ ignore irrelevant rebalancing)

        - *distance* could be $ || w^{reb}_t - w_t || $
