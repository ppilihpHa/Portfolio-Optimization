### Portfolio Basics from the book

- Linear and Log **returns**
    
    - linear return: $r^{lin}_t = \dfrac{p_t - p_{t-1}}{p_{t-1}}$

    - log return: $r^{log}_t = \log p_t - \log p_{t-1}$

    - since returns are rather small the book assumes: $ \mu^{lin}_t \approx \mu^{log}_t $ and therfore $ \Sigma^{lin}_t \approx \Sigma^{log}_t $ 

- **mean vector** and **covariance matrix**

    -  <font color='green'> mean vector </font> $\color{green}{\mu_t}:$

        - $ \mu^{log}_t = \mathrm{E}[r^{log}_t | \mathcal{F}_{t-1} ] $
        - which is basically the expected return of all assets given all information (forecast) up to $t-1$
    
    - <font color='green'> covarianve matrix </font> $\color{green}{\Sigma_t}:$

        - $ \Sigma_t = [ (r^{log}_t - \mu^{log}_t) (r^{log}_t - \mu^{log}_t) | \mathcal{F}_{t-1} ] $
        - which is basically the expected variance of a asset on the diagonal an the expected covariance of two assets on the off-diagonal, given all information (forecast) up to $t-1$

- **Portfolio Return** and **Net Asset Value (NAV)**

    - $w_t$ <font color='red'> portfolio vector, e.q. units of assetes held </font>

        - if units of assets stay constant, then $w_t$ changes with price changes $p_t$

    - $NAV = w^{cap}_t + c^{cap}_t$ <font color='red'> current portfolio value + current cash held  </font>

    - $R^{portfolio}_t = \dfrac{NAV_t - NAV_t-1}{NAV_t-1}$ <font color='red'> currrent return of the portfolio </font>

        - $= w^T_{t-1} * r_t$  <font color='red'> prtfolio vector times individual returns </font>

- first- an second-order **moments** of Potfolios 

    - **Expected portfolio return** $ = w^T_{t-1} \mu $

    - **Portfolio Variance** $ = w^T_{t_1} \Sigma w_{t-1} $ 

    - <font color='red'>Recap:</font> $w_t$ or the weights represent the portfolio (e.g. units held)

- **Cumulative Return**

    - normalized $NAV$ 

- **Transaction Cost**

    - The cost associated with a finacial trade
    
    - *commission fee*: payment to the broker <font color='red'>(dependent on country, broker, etc.)</font>

    - *slippage*: difference between the expected price and the actually executed price <font color='red'>(not necessarily negative)</font>

    - $ (w^{cap}_t \to w^{cap,reb}_t) $ <font color='red'>transaction cost when rebalancing portfolio</font>

    - Therfore the return of the rebalanced portfolio: $ R^{portf}_{t-1} = (w^{reb}_{t-1})^T r_t - (w^{cap}_t \to w^{cap,reb}_t)$

- **Rebalancing**

    - <font color='red'> Trade-off: </font> *frequent rebalancing* to match the currently desired portfolio *vs.* *transaction cost* of rebalancing

    - <font color='red'> Adaptive rebalancing: </font> only rebalance if the *distance* is below some threshold ($\to$ ignore irrelevant rebalancing)

        - *distance* could be $ || w^{reb}_t - w_t || $

- Portfolio **Constraints**

    - <font color='red'>long-only / no-short constraint</font>: if shorting is not allowed: $w \geq 0$ (*no negative weights*)

    - <font color='red'>captial budget constraint</font>: if no other kind of leverage is allowed: $1^Tw \leq 0$ (*no "over"-investing*)

    - <font color='red'>holding constraint</font>: bounds for the portfolio elements: $l \leq w \leq u$ (*no over-/underexposure*)

    - <font color='red'>cardinality constraint</font>: upper bound for active positions: $||w||_0 \leq K$ (*no unrealistic amount of positions*)

    - <font color='red'>turnover constraint</font>: refers to the threshold for rebalancing: $||w - w_0|| \leq u$ (with $w_0$ being the currently held portf.) (*no unecassary rebalancing*)

    - <font color='red'>market neutral-, dollar-neutral-, diversification-, leverage constraint</font>: additional constraints for even more realistic modeling (*not taken into account yet*)

- **performance measures**

    - <font color='red'>(Recap) Expected Return</font>: $ \mathrm{E}[R^{portf}_t] = w^T\mu $ (*just one period t and ignores risk*)

    - <font color='red'>Annualized Return</font>: $ R^{Ann} = \gamma \times w^T\mu$ (*scaling with number of periods $\gamma$ where 252 is typical*)

    - <font color='red'><b>Volatility</b></font>

        - <font color='green'>measure of risk of a portfolio</font>

        - $ Std[R^{portf}_t] = \sqrt{w^T\Sigma w} $

        - Markowitz uses $volatiltiy^2$ i.e. the **variance** of the portfolio

            - $ Var[w^T r_t] = \mathrm{E}[(w^T r_t - w^T \mu)^2] $

            - **very intersting**: Of course the positive part of the variance (i.e. portfolio overperforming the mean return) should not be counted as risk, therefore **downside risk** like the **Semi-Variance** of Markowitz is used (*avoids penalizing return above the mean*) 

        - in order to annualize one must scale with $ \sqrt{\gamma} $ 
    
    - <font color='red'><b>Sharpe Ratio</b></font>

        - <font color='green'>risk adjusted expected excess return</font>

        - $ SR = \dfrac{w^T\mu - r_f}{\sqrt{w^T\Sigma w}} $

        - with $ r_f $ beeing the risk free rate (return of an risk free asset)

        - also annualized with $ \sqrt{\gamma} $ 

        - if one would like to compare the portfolio return not to the return of a risk free asset one could use the **Information Ratio (IR)** which uses another benchmark in the numerator

### Summary of Portfolio Basics Chapter

- A portfolio over $N$ assets is represented by the $N$-dimensional **vector $w$**. The **portfolio return** is given by $w^T r$, where r contains the returns of the assets

    - in practice, returns of assets in the future are unknown, therefore $\mu$ has to be estimated (Thesis potential)

- In practice, portfolio return have to be periodically **rebalanced**, which incurrs **transaction costs**

- Portfolios typically need to satisfy many **constraints**, imposed by regulators or the interest of the investor itself

- The performance of portfolios can measured and monitored in multiple ways, particulary relevant is **Volatility/Variance** and the **Sharpe Ratio**

- There are **heuristic portfolios** that don't need complicated optimization methods but have endured the test of time

- **Risk based portfolios**, such as the **GMVP** (Thesis) focus on reducing the risk (variability in the returns)

    - as mentioned above, the variability of future returns is also unknown and has to be estimated (Ledoit, Thesis potential)
